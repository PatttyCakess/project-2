```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load (Remember to Change These)
path = "toronto_weather_data.csv"
```

```python
# Read Data File 
weather_data = pd.read_csv(path)
```

```python
# store into Pandas DataFrames
weather_df = pd.DataFrame(weather_data)
```

```python
#filter for just Toronto data 
toronto_weather = weather_df.loc[weather_df["NAME"] == 'TORONTO CITY, ON CA']
toronto_weather
```

```python
#drop unnecessary columns 
toronto_weather_clean = toronto_weather.drop(columns=['STATION', 'LATITUDE', 'LONGITUDE',
                                                      'ELEVATION', 'DAPR', 'MDPR', 'SNOW',
                                                      'WDFG', 'WESD', 'WESF', 'WSFG'])
toronto_weather_clean
```

```python
#check data types 
toronto_weather_clean.dtypes
```

```python
#change data types 
toronto_weather_clean['NAME'] = toronto_weather_clean['NAME'].astype('string')
toronto_weather_clean['DATE'] = pd.to_datetime(toronto_weather_clean['DATE'])
toronto_weather_clean['DATE'] = toronto_weather_clean['DATE'].dt.strftime('%m/%d/%Y')
```

```python
#check to make sure changes were successful 
toronto_weather_clean.dtypes
```

```python
#rename columns
toronto_weather_clean.rename(columns={'NAME': 'CITY','SNWD': 'SNOW_DEPTH', 'TAVG': 'AVG_TEMP',
                                      'TMAX': 'MAX_TEMP', 'TMIN': 'MIN_TEMP' }, inplace=True)
toronto_weather_clean
```

```python
# Write the data to a CSV
toronto_weather_df = '../toronto_weather_df.csv'
toronto_weather_clean.to_csv(toronto_weather_df, index=False)
```
